# Calculating the churn percentage area-wise

```python
100*df.groupby(['Area code'])['Churn'].value_counts()/len(df)
# The churn percentage for 415 code is 7 but the ratio is almost same.
```



*   Plotting the graph between churn and Area code.



```python
plt.figure(figsize=(10,5))
df.groupby(['Area code'])['Churn'].value_counts().plot.bar()
plt.ylabel('Value_counts',color='r',fontsize=10)
plt.title('Bar graph',color='r',fontsize=15)
```

#Plotting dist plot for numerical column.

```python
df.info()
```

```python
nummerical_column=['Number vmail messages','Total day minutes','Total day calls','Total day charge','Total eve minutes','Total eve calls','Total eve charge','Total night minutes','Total night calls','Total night charge','Total intl minutes','Total intl calls','Total intl charge','Customer service calls']
```

```python
for i in nummerical_column:
  plt.figure(figsize=(10,5))
  sns.distplot(df[i])
  plt.title(f'Distribution_plot for {i}',color='r')
  plt.show()  
```

```python
# As we see that the data is normally distributed for Total day minute,Total day calls,Total day charge,Total eve minutes,total eve calls,Total eve charge,Total night minute,total night calls, Total night charge,Total int minutes,Total int charge,
```

# Data Manipulation

```python
df['total_monthly_charges']=df['Total day charge']+df['Total eve charge']+df['Total night charge']+df['Total intl charge']
# Creating new column to dataset i.e total_monthly_charges
```

```python
df['total_monthly_minute']=df['Total day minutes']+df['Total eve minutes']+df['Total night minutes']+df['Total intl minutes']
#Creating new column to dataset i.e total_monthly_minute
```

#Plotting the count plot between different parameters.

```python
list_of_parameters=list(df.columns)
```

```python
for i in list_of_parameters:
  plt.figure(figsize=(40,8))
  sns.countplot(df,x=i,hue='Churn')
  plt.title(f'count plot between {i} and Churn',fontsize=20)
  plt.show()
```
